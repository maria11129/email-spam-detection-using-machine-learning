```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from imblearn.over_sampling import SMOTE  
```

```python
df = pd.read_csv('spam.csv', encoding='latin1')
```

```python
df.rename(columns={'v1': 'Category', 'v2': 'Message'}, inplace=True)
print(df)
```

```python
data = df.where(pd.notnull(df), '')
```

```python
data.loc[data['Category'] == 'spam', 'Category'] = 0
data.loc[data['Category'] == 'ham', 'Category'] = 1
```

```python
X = data['Message']
Y = data['Category']
```

```python
print(X)
```

```python
print(Y)
```

```python
df['Category'] = encoder.fit_transform(df['Category'])
```

```python
# drop last 3 cols
df.drop(columns=['Unnamed: 2','Unnamed: 3','Unnamed: 4'],inplace=True)
```

#data is imbalanced

```python
df['num_characters'] = df['Message'].apply(len)
```

```python
df.head()
```

```python
# Basic tokenization by whitespace
df['num_words'] = df['Message'].apply(lambda x: len(x.split()))
```

```python
df.head()
```

```python
# Approximate sentence splitting based on periods
df['num_sentences'] = df['Message'].apply(lambda x: len(x.split('.')))
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=3)
```

```python
feature_extraction = TfidfVectorizer(min_df=1, stop_words='english', lowercase=True)
```

```python
X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)
```

```python
Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
```

```python
smote = SMOTE(random_state=42)  
X_train_resampled, Y_train_resampled = smote.fit_resample(X_train_features, Y_train)
```

```python
model = LogisticRegression()
model.fit(X_train_resampled, Y_train_resampled)
```

```python
prediction_on_training_data = model.predict(X_train_resampled)
accuracy_on_training_data = accuracy_score(Y_train_resampled, prediction_on_training_data)
```

```python
prediction_on_test_data = model.predict(X_test_features)
accuracy_on_test_data = accuracy_score(Y_test, prediction_on_test_data)
```

```python
print(f"Accuracy on training data: {accuracy_on_training_data * 100:.2f}%")
print(f"Accuracy on test data: {accuracy_on_test_data * 100:.2f}%")
```

```python

input_your_mail =["hello uve won a gift of 1000 dollars"]
input_data_features = feature_extraction.transform(input_your_mail)
```

```python
prediction = model.predict(input_data_features)

print(prediction)

if prediction[0] == 1:
    print('Ham mail')
else:
    print('Spam mail')
```
